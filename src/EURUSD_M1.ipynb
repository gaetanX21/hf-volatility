```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

sns.set_theme()
```

```python
df = pd.read_csv('../datasets/new_EURUSD/EURUSD_M1.csv', sep='\t')
# rename "Time" to "Timestamp"
df = df.rename(columns={'Time': 'Timestamp'})
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df['Date'] = df['Timestamp'].dt.date
df['Time'] = df['Timestamp'].dt.time
df['WeekDay'] = df['Timestamp'].dt.dayofweek
df['Hour'] = df['Timestamp'].dt.hour

df['Minute'] = df['Timestamp'].dt.minute
```

```python
start_day = pd.to_datetime('2023-12-11')
end_day = pd.to_datetime('2024-03-08')

df = df[(df['Timestamp'] >= start_day) & (df['Timestamp'] <= end_day)]
```

```python
df.groupby('Date').size().plot(kind='bar')
```
